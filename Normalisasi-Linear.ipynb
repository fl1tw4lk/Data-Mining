```python
import numpy as np

# Dataset
data = np.array([
    [0, 2, 1],
    [1, 7, 1],
    [2, 6, 3],
    [5, 1, 4],
    [3, 3, 4]
])

# Function for Linear normalization
def z_score_normalization(data):
    print(r"""
        ┌┐┌┌─┐┬─┐┌┬┐┌─┐┬  ┬┌─┐┌─┐┌─┐┬  ┬  ┬┌┐┌┌─┐┌─┐┬─┐
        ││││ │├┬┘│││├─┤│  │└─┐├─┤└─┐│  │  ││││├┤ ├─┤├┬┘
        ┘└┘└─┘┴└─┴ ┴┴ ┴┴─┘┴└─┘┴ ┴└─┘┴  ┴─┘┴┘└┘└─┘┴ ┴┴└─
    """)
    print("Data Asli:")
    print(data)
    
    # Calculate mean and standard deviation for each feature
    means = np.mean(data, axis=0)
    std_devs = np.std(data, axis=0, ddof=1)
    
    # Step-by-step explanation for calculating mean and standard deviation
    # Display mean and standard deviation for each feature
    print("-"*10,"Mean (x̄) untuk setiap fitur:","-"*10)
    for j in range(data.shape[1]):
        print(f"Fitur {j+1}: {means[j]:.4f}")
    print("\nLangkah-langkah perhitungan mean:")
    # Calculate and show mean for each feature
    for j in range(data.shape[1]):
        feature_data = data[:, j]
        mean = np.mean(feature_data)
        print(f"\nFitur {j+1}:")
        print(f"  Data: {feature_data}")
        print(f"  x̄ = (Hasil Jumlah data) / (Banyak Data):")
        print(f"  Mean (x̄) = {np.sum(feature_data)} / {len(feature_data)} = {mean:.4f}")
    print("="*70)
    print("-"*10,"Standard Deviation (σ) untuk setiap fitur:","-"*10)
    for j in range(data.shape[1]):
        print(f"Fitur {j+1}: {std_devs[j]:.4f}")
    
    # Calculate and show standard deviation for each feature
    print("\nLangkah-langkah perhitungan deviasi standar:")
    for j in range(data.shape[1]):
        feature_data = data[:, j]
        mean = np.mean(feature_data)
        squared_diff = (feature_data - mean) ** 2
        std_dev = np.sqrt(np.sum(squared_diff) / (len(feature_data) - 1))  # Sample std deviation (ddof=1)
        print(f"\nFitur {j+1}:")
        print(f"  Data: {feature_data}")
        print(f"  Deviasi Standar (σ) = sqrt(Sum of (Data - Mean)²) / (Jumlah Data - 1)")
        print(f"  σ = sqrt({np.sum(squared_diff)} / {len(feature_data) - 1}) = {std_dev:.4f}")
    
    print("="*70)
    # Calculate Linear normalization for each element
    normalized_data = (data - means) / std_devs
    
    # Display step-by-step Linear calculations
    print("-"*10,"Langkah-langkah perhitungan normalisasi Linear:","-"*10)
    for j in range(data.shape[1]):
        print(f"\nFitur {j+1}:")
        for i in range(data.shape[0]):
            z_value = (data[i, j] - means[j]) / std_devs[j]
            print(f"  x̄ {i+1}{j+1}: {data[i, j]} -> ({data[i, j]} - {means[j]:.4f}) / {std_devs[j]:.4f} = {z_value:.4f}")
    
    # Display normalized data
    print("\nData setelah normalisasi:")
    print(np.round(normalized_data, 4))
    return normalized_data

# Run the normalization
normalized_data = z_score_normalization(data)
```

```python
import numpy as np

# Dataset
data = np.array([
    [10, 12, 2],
    [8.8, 8, 4],
    [4.3, 3, 7],
    [6.4, 5, 5],
    [10.2, 9, 3],
    [13, 16, 1],
    [5.672, 5, 6]
])



# Fungsi untuk normalisasi Min-Max ke rentang [-1, 1]
def normalize_min_max(data):
    # Menampilkan data asli
    print("="*10, "Normalisasi Min-Max [-1,1]", "="*10)
    print("Data Asli:")
    print(data)
    
    # Menentukan min dan max untuk setiap fitur
    min_vals = np.min(data, axis=0)
    max_vals = np.max(data, axis=0)
    
    # Menampilkan nilai min dan max untuk setiap fitur
    print("\nNilai minimum untuk setiap fitur:")
    print(min_vals)
    print("\nNilai maksimum untuk setiap fitur:")
    print(max_vals)
    
    # Menghitung normalisasi untuk setiap elemen ke rentang [-1, 1]
    normalized_data = 2 * ((data - min_vals) / (max_vals - min_vals)) - 1
    
    # Menampilkan langkah-langkah perhitungan normalisasi
    print("\nLangkah-langkah perhitungan normalisasi:")
    for j in range(data.shape[1]):  # Untuk setiap fitur (kolom)
        print(f"\nFitur {j+1}:")
        for i in range(data.shape[0]):  # Untuk setiap data point (baris)
            norm_value = 2 * (data[i, j] - min_vals[j]) / (max_vals[j] - min_vals[j]) - 1
            print(f"  x̄ {i+1}{j+1}: {data[i, j]} -> ({data[i, j]} - {min_vals[j]}) / ({max_vals[j]} - {min_vals[j]}) = {norm_value:.4f}")
            
    # Menampilkan data yang telah dinormalisasi
    print("\nData setelah normalisasi:")
    print(np.round(normalized_data, 4))
    return normalized_data

# Menjalankan normalisasi Min-Max ke rentang [-1, 1]
normalized_data = normalize_min_max(data)
```
